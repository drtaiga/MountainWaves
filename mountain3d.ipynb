```python
%matplotlib notebook
import matplotlib.pyplot as plt
import mountain_wave_model as mw_model
import numpy as np
from IPython.display import display, Math, Latex
from pylab import rcParams
rcParams['figure.figsize'] = 12, 6.5
```

```python
# Create input parameter objects for each run. By default, each object holds 
# the default values for the input parameters. Here we'll change the  
# Brunt-Väisälä frequencies from their default values. We'll also change a few other
# selected parameters.

run0_input = mw_model.input_params_3d() # Create the object.
run0_input.N_l=0.125e-2   # Define upper and lower layer Brunt-Väisälä frequencies (1/s)
run0_input.N_u=0.125e-2
run0_input.U_l=15.0     # Upper and lower layer wind velocities (m/s)
run0_input.U_u=15.0
run0_input.h=0.5e3
run0_input.a=5.0e3
fn = run0_input.h* run0_input.N_l /run0_input.U_l
run0 = mw_model.mountain_wave_3d(run0_input)
run0.run()

run1_input = mw_model.input_params_3d()
run1_input.N_l=0.25e-2
run1_input.N_u=0.25e-2
run1_input.U_l=15.0 # Upper and lower layer Brunt-Väisälä frequencies
run1_input.U_u=15.0
run1_input.h=0.5e3
run1_input.a=5.0e3
run1 = mw_model.mountain_wave_3d(run1_input)
run1.run()

run2_input = mw_model.input_params_3d()
run2_input.N_l=0.5e-2
run2_input.N_u=0.5e-2
run2_input.U_l=15.0 # Upper and lower layer Brunt-Väisälä frequencies
run2_input.U_u=15.0
run2_input.h=0.5e3
run2_input.a=5.0e3
run2 = mw_model.mountain_wave_3d(run2_input)
run2.run()
```

```python
plt.figure(figsize=(12,6.5))
zz=mw_model.comparison_plot(run0.XY,run0.YX,
                         run0.XZ,run0.ZX,
                         run0.YZ,run0.ZY,
                         run0.x, run0.y, run0.z,
                         run0.h_3d,run1.h_3d,run2.h_3d,
                         0.,0.,10.,
                         run2_input.h, run2_input.a)
zz.do_plot()
plt.show()
```

```python
vvm=np.max([np.max(run0.precipitation),np.max(run1.precipitation), np.max(run2.precipitation)])*3600.
plt.figure(figsize=(6, 10))
plt.subplot(311)
plt.pcolormesh(run0.XY/1000,run0.YX/1000,run0.precipitation*3600.,vmax=vvm)
plt.colorbar()
plt.xlim(-50,50)
plt.ylim(-25,25)
#plt.show()
plt.subplot(312)
plt.pcolormesh(run1.XY/1000,run1.YX/1000,run1.precipitation*3600.,vmax=vvm)
plt.colorbar()
plt.xlim(-50,50)
plt.ylim(-25,25)
#plt.show()
plt.subplot(313)
plt.pcolormesh(run2.XY/1000,run2.YX/1000,run2.precipitation*3600.,vmax=vvm)
plt.colorbar()
plt.xlim(-50,50)
plt.ylim(-25,25)
plt.show()
```
